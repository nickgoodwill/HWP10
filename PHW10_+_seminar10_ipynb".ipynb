Создать новый столбец в таблице с
пингвинами, который будет отвечать за
показатель длины клюва пингвина.
high - длинный(от 42)
middle - средний(от 35 до 42)
low - маленький(до 35)


```python
penguins.loc[penguins['bill_length_mm'] <= 35, 'beak_length'] = 'маленький'
penguins.loc[(penguins['bill_length_mm'] > 35) & (penguins['bill_length_mm'] <= 42),
'beak_length'] = 'средний'
penguins.loc[penguins['bill_length_mm'] > 42, 'beak_length'] = 'длинный'

penguins.head
```

```python
import random
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
data.head()
```

```python
data['tmp'] = 1
data.set_index([data.index, 'whoAmI'], inplace=True)
data = data.unstack(level=-1, fill_value = 0).astype(int)
data.columns = data.columns.droplevel()
data.columns.name = None
print(data)
```
